Beginner Level Task4:Image to Pencil Sketch with Python

IMPORTING REQUIRED LIBRARIES

```python

import matplotlib.pyplot as pt
import cv2
```

Accessing The Original image

```python
img=cv2.imread("noora.jpg")
pt.imshow(img)
```

Now,we will convert BGR(blue green red) image to RGB(red green blue)

```python
r_image=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
pt.imshow(r_image)
```

Converting RGB image to Gray Scale

```python
gr_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
pt.imshow(gr_img)
```

To invert the Gray Scale Image

```python
inv_img=cv2.bitwise_not(gr_img)
pt.imshow(inv_img)
```

To Blur the image using Gaussian Function

```python
blur_image=cv2.GaussianBlur(inv_img,(21,21),0)
pt.imshow(blur_image)
```

Inverted Blurry Image

```python
inv_blur_img=cv2.bitwise_not(blur_image)
pt.imshow(inv_blur_img)
```

The Sketch can be generated by performing bit-wise division on the grayscale image and the inverted-blurred image

```python
sket_img=cv2.divide(gr_img,inv_blur_img,scale=256.0)
```

To save the sketch image

```python
cv2.imwrite('sket.png', sket_img)
```

To display original image and sketch image

```python
pt.figure(figsize=(14,8))
pt.subplot(1,2,1)
pt.title('Original_image', size=18)
pt.imshow(r_image)
pt.axis('off')

pt.subplot(1,2,2)
pt.title('Sketch_image', size=18)
rgb_sketch=cv2.cvtColor(sket_img, cv2.COLOR_BGR2RGB)
pt.imshow(rgb_sketch)
pt.axis('off')
pt.show()
```
